# 第4章　例外値への対応

## 4.1 欠損値の表現とその確認方法

```python
import pandas as pd
import numpy as np

df = pd.read_csv('sample_4_1.csv')

df
```

```python
df.info()
```

```python
df.isna()   # df.isnull()でも同一
```

```python
df.isna().sum(axis=0)
```

```python
df.isna().sum(axis=1)
```

```python
df_replace = df.replace('NAN', np.nan)

df_replace
```

```python
df_replace = df.replace({
    'NAN': np.nan,
    'na': np.nan
})

df_replace
```

```python
df.replace(
    ['NAN', 'na'],
    [np.nan, np.nan]
)

df_replace
```

```python
df_replace = df.replace(r'NAN|na|Null', np.nan, regex=True)

df_replace
```

```python
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5, 6],
    'B': [7, 8, 9, 10, 11, 12]
})

df
```

```python
df.info()
```

```python
df.loc[1, 'B'] = None
df.loc[2, 'B'] = np.nan
df.loc[3, 'B'] = pd.NA
df.loc[4, 'B'] = float('nan')

df
```

```python
df.info()
```

```python
print(np.isnan(df.loc[1, 'B']))
print(df.loc[1, 'B'].dtype)
```

```python
df_float32 = df.astype('float32')

df_float32.info()
```

```python
print(np.isnan(df_float32.loc[1, 'B']))
print(df_float32.loc[1, 'B'].dtype)
```
