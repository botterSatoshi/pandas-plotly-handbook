## 3.1 値の演算

```python
import pandas as pd
import numpy as np

def load_3_1():
    df = pd.read_csv(
        'sample_3_1.csv',
        dtype={
            'val1': 'float64',
            'val2': 'float64'
        },
        index_col=0
    )
    return df

df = load_3_1()
df
```

```python
df.iat[0, 0] += 2.
df.iat[1, 0] -= 2.
df.iat[2, 0] *= 2.
df.iat[3, 0] /= 2.

df.iat[0, 1] %= 2.
df.iat[1, 1] //= 2.
df.iat[2, 1] **= 2.

df.head()
```

```python
df['val1'] *= 1000
df.iloc[1:4, 1] /= 1000

df.head()
```

```python
df = load_3_1()
```

```python
%%timeit

for i in range(len(df)):
    df.iloc[i, 0] /= 1000
```

```python
df = load_3_1()
```

```python
%%timeit

df.iloc[:, 0] /= 1000
```

```python
df.iat[0, 0] = 0
df.iloc[1:4, 1] = 0

df.head()
```

```python
df.iloc[1:4, :] = np.array([
    [1, 2], [3, 4], [5, 6]
])

df.head()
```

```python
df = load_3_1()
boolean_mask = df['val1'] < 150

boolean_mask
```

```python
df_mask = df.mask(boolean_mask, -1)

df_mask
```

```python
df_mask = df.mask(boolean_mask)

df_mask
```

```python
df_where = df.where(boolean_mask, -1)   # df.mask(~boolean_mask, -1)と同じ

df_where
```

```python
df = load_3_1()

df['feat1'] = df['val1'] * df['val2']

df.head()
```

```python
df['feat2'] = np.log(df['val2'])

df.head()
```

```python
df['feat3'] = df['val1'].shift(1)
df['feat4'] = df['val1'].shift(2)

df.head()
```

```python
df['feat5'] = df['val1'] - df['val1'].shift(1)

df.head()
```

```python
df['feat6'] = df['val2'].cumsum()

df.head()
```

```python
df = load_3_1()
df['smooth_val1'] = df['val1'].rolling(3).mean()
df['smooth_val2'] = df['val2'].rolling(3, center=True).mean()

df
```

```python
df = load_3_1()
df_agg = df.agg(['sum', 'mean', 'median', 'min', 'max'])

df_agg
```
